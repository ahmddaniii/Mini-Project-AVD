## **Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Data tidak memiliki outliers

# **Construct Data**

Menambahkan kolom spending category dan kolom costumer segmen, adapun tujuan dari penambahan dua kolom tersebut adalah:
*   "Spending Category" membantu memahami pola pengeluaran pelanggan dan strategi pemasaran berdasarkan daya beli.
*   "Customer Segment" memungkinkan bisnis menyesuaikan strategi pemasaran, meningkatkan loyalitas, dan mengoptimalkan konversi pelanggan.




```python
df["Spending Category"] = pd.cut(df["Purchase Amount (USD)"],
                                 bins=[0, 50, 200, 500, 1000, float("inf")],
                                 labels=["Low", "Medium", "High", "Premium", "Luxury"])

df["Customer Segment"] = "Regular"
df.loc[df["Spending Category"] == "High", "Customer Segment"] = "High-Value"
df.loc[df["Spending Category"] == "Low", "Customer Segment"] = "New Customer"
```

```python
df
```

# **Data Reduction**

Pada dataset ini saya menghapus beberapa kolom seperti kolom category, payment method, location, shipping type, season, discount applied, promo code used, previous purchased, dan preferred payment method.

```python
df = df.drop(columns=["Category", "Payment Method", "Location", "Shipping Type", "Season", "Discount Applied", "Promo Code Used", "Previous Purchases", "Preferred Payment Method"])
```

## **Visualisasi Data**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Distribusi Kategori Pengeluaran**

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Spending Category", hue="Spending Category",palette="coolwarm", order=df["Spending Category"].value_counts().index)
plt.title("Distribusi Kategori Pengeluaran")
plt.xlabel("Kategori Pengeluaran")
plt.ylabel("Jumlah Pelanggan")
plt.show()
```

Insight dan tindakan yang dapat diambil:

*   Mayoritas pelanggan berada dalam kategori pengeluaran rendah dan menengah.
*   Tindakan: Fokus pada strategi upselling untuk pelanggan di kategori menengah agar mereka meningkatkan pembelian ke kategori tinggi.



## **Frekuensi Pembelian Pelanggan**

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Frequency of Purchases",hue="Frequency of Purchases", palette="viridis", order=df["Frequency of Purchases"].value_counts().index)
plt.title("Frekuensi Pembelian Pelanggan")
plt.xlabel("Frekuensi Pembelian")
plt.ylabel("Jumlah Pelanggan")
plt.xticks(rotation=45)
plt.show()
```

Insight dan tindakan yang dapat diambil:


* Sebagian besar pelanggan melakukan pembelian mingguan dan dua mingguan.
* Tindakan: Berikan diskon dan penawaran eksklusif bagi pelanggan yang berbelanja lebih sering untuk meningkatkan retensi.



## **Produk yang Paling Sering Dibeli**


```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y="Item Purchased", hue="Item Purchased",palette="mako", order=df["Item Purchased"].value_counts().index)
plt.title("Produk yang Paling Sering Dibeli")
plt.xlabel("Jumlah Pembelian")
plt.ylabel("Produk")
plt.show()
```

Insight dan tindakan yang dapat diambil:
* Beberapa produk lebih diminati dibanding lainnya, seperti blouse, sweater, dan jeans.
* Tindakan: Fokuskan kampanye pemasaran pada produk-produk populer dan gunakan strategi cross-selling dengan produk terkait.

## **Segmentasi Pelanggan Berdasarkan Pengeluaran**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Customer Segment", y="Purchase Amount (USD)", hue="Customer Segment", palette="Set2")
plt.title("Segmentasi Pelanggan Berdasarkan Pengeluaran")
plt.xlabel("Segmentasi Pelanggan")
plt.ylabel("Jumlah Pembelian (USD)")
plt.xticks(rotation=45)
plt.show()
```

Insight dan tindakan yang dapat diambil:
* Pelanggan di segmen VIP dan Regular cenderung memiliki nilai pembelian lebih tinggi dibanding segmen lainnya.
* Tindakan:
Memberikan program loyalitas dan reward untuk pelanggan VIP agar mereka tetap setia.
Tawarkan diskon awal atau paket bundling untuk pelanggan baru agar mereka lebih banyak bertransaksi.
